```python
def word2int(textnum, numwords={}):
        found_words = []
        firstNumber = []
        fractionNumber = "" 
        current = result = lastunit = 0
        dist = 0
        
    #if not numwords:
        units = [
            "cero", "uno", "dos", "tres", "cuatro", "cinco", "seis", "siete", "ocho",
            "nueve", "diez", "once", "doce", "trece", "catorce", "quince",
            "dieciseis", "diecisiete", "dieciocho", "diecinueve",
        ]
        
        tens = ["", "", "veinte", "treinta", "cuarenta", "cincuenta", "sesenta", "setenta", "ochenta", "noventa"]
        
        scales = ["cien", "mil", "millon", "billon", "trillon"]
        
        fractions = [
            "", "", "medio", "tercio", "cuarto"
        ]
        
        ordinal_endings = [('cientos', 100)]
        word_endings = []#[('s', "")]

        numwords["y"] = (1, 0, False)
        for idx, word in enumerate(units):       numwords[word] = (1, idx, False)
        for idx, word in enumerate(tens):        numwords[word] = (1, idx * 10, False)
        for idx, word in enumerate(scales):      numwords[word] = (1, 10 ** (idx * 3 or 2), False)
        for idx, word in enumerate(fractions):   numwords[word] = (0, idx, True)

       
        words = textnum.split()
        numbers = []
        
        for index, word in enumerate(words):
            if "un" in word or "una" in word:
                word = "uno"
            if "media" in word:# or "media" in word:
                word = "medio"
            
            # TODO => veintiuno, novecientos, quinientos, fracciones
            
            # separa el "cientos"
            for ending, replacement in ordinal_endings:
                if word.endswith(ending):
                    lastunit = replacement
                    word = word[:-len(ending)]
                    
            if word not in numwords:
                continue #return found_words, result + current
            else:
                if len(found_words) == 0:
                    firstNumber = [ index, word ]
                found_words.append(word)
            
            if len(firstNumber) > 0:
                if index > firstNumber[0] and index - firstNumber[0] > 2:
                    return found_words, result + current
            
            scale, increment, fraction = numwords[word]
            
            if fraction:
                if current == 0: 
                    current = 1
                fractionNumber = str(current) + "/" + str(increment)
                    #break
                return found_words, fractionNumber

            if lastunit != 0:
                result += current + increment * lastunit
                current = 0
                lastunit = 0
            
            elif increment >= 1000:
                if current == 0: current = 1
                current *= increment
                increment = 0 
                
            else:
                current = current * scale + increment
            
            #print(index, word, found_words, result + current, firstNumber)

        return found_words, result + current
```

```python
def wsd_caracteristicas_colocacion(context, instance, pos, dist=2):
    features = {}
    
    # Se obtienen los atributos de la instancia
    #pos = instance.position
    con = context

    # Las palabras serán almacenadas de acuerdo a su posición.
    prev_words = []
    next_words = []

    # A partir de una posición dada, se obtienen las palabras previas requeridas
    # dependiendo la distancia.
    for i in range(max(0, pos-dist), pos):
        prev_words.append(con[i])
    features["previous"] = (' '.join(prev_words))

     # A partir de una posición dada, se obtienen las palabras posteriores requeridas
    # dependiendo la distancia.
    for i in range(pos+1, min(pos+dist+1, len(con))):
        next_words.append(con[i])
    features['next'] = (' '.join(next_words))
      
    return features
```

```python
stopwords = [".", ","]

#new_tokenized_words = [word for word in enumerate(tokenized_words) if word not in stopwords]
new_tokenized_words = [word[1] for word in token_words_set] #if word not in stopwords]
new_tokenized_words
```

```python
ingredients = ["chiles", "cebolla", "crema", "elote", "consomé", "mantequilla"]

ngrams_ingredients = []

# Search by Ingredients
for items in token_words_set:
    pos = items[0]
    words = items[1]
    
    for ingredient in ingredients:
        if ingredient in words:
            item = { ingredient: [wsd_caracteristicas_colocacion(new_tokenized_words, ingredient, pos, dist=4), pos] } 
            ngrams_ingredients.append(item)

ngrams_ingredients
```

```python
# Search by measures
ngrams_measures = []

for items in token_words_set:
    pos = items[0]
    words = items[1]
    
    for measure in measures:
        if measure == words:
            item = { measure: [ wsd_caracteristicas_colocacion(new_tokenized_words, measure, pos, dist=4), pos] }
            ngrams_measures.append(item)

ngrams_measures
```

```python
ingredients_with_measures = {}

for i, measures in enumerate(ngrams_measures):
    for m in measures:
        prev_words = measures[m][0]["previous"]
        next_words = measures[m][0]["next"]
        measure_idx = measures[m][1] + 1
        
        text = prev_words + " " + next_words
        words_num, integer = word2int(text)
            
        for idx, words in enumerate(next_words.split()):
            for ingredient in ingredients:
                if ingredient == words and integer != 0:
                    if ingredient not in ingredients_with_measures:
                        ingredients_with_measures[ingredient] = [ integer, m, measure_idx + idx]
                    #if found_item not in ingredients_with_measures:
                    #ingredients_with_measures.append(found_item)
            
ingredients_with_measures
```

```python
# Search by ingredients

ingredients_with_quantity = []

for ingredients in ngrams_ingredients:
    for i in ingredients:
        prev_words = ingredients[i][0]["previous"]
        next_words = ingredients[i][0]["next"]
        index = ingredients[i][1]
        
        text = prev_words + " " + next_words
      
        words_num, integer = word2int(text)
        
        # get index of words_num in 
        
        if integer != 0:
            found_item = [ i, integer, "", index]
        else:
            found_item = [ i, "", "", index ]
        
        if found_item not in ingredients_with_quantity:
            ingredients_with_quantity.append(found_item)
                    
ingredients_with_quantity
```

```python
only_ingredients = []

for iq in ingredients_with_quantity: 
    ing_quantity = iq[1]
    if ing_quantity == '':
        only_ingredients.append(iq)

only_ingredients
```

```python
final_ingredients = {}

# Agregar ingredientes con medidas y cantidades
for im in ingredients_with_measures.items():
    ingredient_m = im[0]
    quantity_m = im[1][0]
    measure_m = im[1][1]
    index_m = im[1][2]
    
    for iq in ingredients_with_quantity:
        ingredient_q = iq[0]
        ing_quantity = iq[1]
        
        # Ingredientes con cantidad y medidas
        if ingredient_m == ingredient_q: 
           # if im not in final_ingredients:
            final_ingredients[ingredient_m] = [ quantity_m, measure_m, index_m ] #optimizar


# Agregar solo ingredientes con cantidades
for iq in ingredients_with_quantity:
    ingredient_q = iq[0]
    quantity_q = iq[1]
    
    if ingredient_q not in final_ingredients and quantity_q != '':
        iq_copied = iq.copy()
        iq_copied.remove(ingredient_q)
        
        final_ingredients[ingredient_q] = iq_copied

# Agregar solo ingredientes
for oi in only_ingredients:
    only_ing = oi[0]
    
    if only_ing not in final_ingredients:
        oi_copied = oi.copy()
        oi_copied.remove(only_ing)
        final_ingredients[only_ing] = oi_copied 
 

final_ingredients 
        
```

```python
# Obtener adjetivos
final_ingredients_adj = {}

for v_ingredient in final_ingredients.values():
    adj = []
    v_ingredient_idx = v_ingredient[2]
    
    ingredient = token_words_set[v_ingredient_idx][1]
    
    for t_word in token_words_set:
        t_word_idx = t_word[0]
        t_word_word = t_word[1]
        t_word_tag = t_word[3]
    
        if t_word_idx == v_ingredient_idx:
            for idx in range(v_ingredient_idx + 1, len(token_words_set)): # cambiar a infinito
                searched_adj = token_words_set[idx]
                searched_adj
                adj_word = searched_adj[1]
                adj_tag = searched_adj[3]
            
                if adj_tag == "ADJ":
                    adj.append(adj_word)
                elif adj_tag != "ADJ":
                    break
                    
    final_ingredients_adj[ingredient] = adj      

final_ingredients_adj
```
